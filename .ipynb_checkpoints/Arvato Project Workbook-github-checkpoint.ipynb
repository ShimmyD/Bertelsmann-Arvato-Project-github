# Capstone Project: Create a Customer Segmentation Report for Arvato Financial Services

In this project, we will analyze demographics data for customers of a mail-order sales company in Germany anddemographics information for the general population. The datasets used have been provided by Bertelsmann Arvato Analytics, and represents a real-life data science task.

## Problems to be solved
1.	Identify parts of the general population that are more likely to be part of the company's main customer base
2.	Make prediction on whether individuals will respond to company’s campaign.

## Strategies and Metrics

We'll use unsupervised learning techniques to perform customer segmentation for the first problem. By comparing percentage of established customer with percentage of general population in each segment, we locate the group of general population that should be targeted in the mail-order sale campaign.

For second problem we will train a ML model with a large dataset of demographics info for the campaign targets to predict individual’s responses. The dataset has a large output class imbalance, where most individuals did not respond to the mailout. Thus, ROC_AUC is used as evaluation metric.

## Problem 1: Customer Segmentation Report

Here we will use unsupervised learning techniques to describe the relationship between the demographics of the company's existing customers and the general population of Germany. We will discover parts of the general population that are more likely to be part of the mail-order company's main customer base, and which parts of the general population are less so.

### Data Pre-processing

```python
def identify_nulls(combined):
    """
    identify variable types. Replace -1,0 and 9 with np.nan
    
    INPUT:
        combined: a concatenated dataframe of azdias and customers datasets
    OUTPUT:
        combined: a dataframe with NaNs representing all unknowns
        numerics: a list of numeric variables
        categoricals: a list of categorical variables
        
    """
    # import a csv file that map variables to variable types (binary, categorical and numeric)
    var_known=pd.read_csv('data/variables known.csv')
    var_known=var_known.loc[var_known.Variables.isin(combined.columns),:]
    combined=combined.loc[:,combined.columns.isin(var_known.Variables.tolist())]

    # find numeric and categorical variables according to DIAS Information Levels - Attributes 2017.xlsx
    numerics=var_known.loc[var_known.Type.isin(['numeric','ordinal','binary']),'Variables'].tolist()
    categoricals=var_known.loc[var_known.Type=='non-ordinal','Variables'].tolist()

    # replace -1 , 0, 9 that are representing unknown to nan. 
    unknowns_0=pd.read_csv('data/unknowns_0.csv')  # a list of variables with 0 representing unknows
    unknowns_negone=pd.read_csv('data/unknowns_-1.csv')  # a list of variables with -1 representing unknows
    unknowns_9=pd.read_csv('data/unknowns_9.csv')  # a list of variables with 9 representing unknows

    unknowns_0=np.intersect1d(unknowns_0.Variables,var_known.Variables)
    unknowns_negone=np.intersect1d(unknowns_negone.Variables,var_known.Variables)
    unknowns_9=np.intersect1d(unknowns_9.Variables,var_known.Variables)

    combined[unknowns_0]=combined[unknowns_0].replace(0,np.NaN)
    combined[unknowns_negone]=combined[unknowns_negone].replace(-1,np.NaN)
    combined[unknowns_9]=combined[unknowns_9].replace(9,np.NaN)
    combined.CAMEO_DEUG_2015=combined.CAMEO_DEUG_2015.replace('X',np.nan).astype('float')
    combined['CAMEO_INTL_2015']=combined['CAMEO_INTL_2015'].replace('XX',np.nan).astype('float')
#     combined['CAMEO_DEU_2015']=combined['CAMEO_DEU_2015'].replace('XX',np.nan)
    
    return combined, numerics, categoricals
```

```python
def identify_cols_to_drop(combined,clustering=True):
    """ Identify cols and rows with missing values over 70%, if classification, not clustering, drop trainset rows and cols 
    and return a list of cols to drop for testset. if clustering, drop cols and rows.
        return a list of cols to drop. 
        
        INPUT:
            combined: a dataframe with NaNs representing all unknowns
            trainset: boolean. check if it's trainset
        OUTPUT:
            combined: a dataframe after dropping rows and cols
            cols_drop: a list of cols to drop for test set
    """
    # drop rows with >70% missing values
    combined=combined[combined.isnull().sum(axis=1)/combined.shape[1]<0.7]
    print(combined.shape)
    # drop columns with >70% missing values
    cols_drop=combined.columns[combined.isnull().sum()/len(combined)>0.7]
    combined=combined.drop(columns=cols_drop)
    
    if clustering:  # verify if it's test set or train set, if trainset, drop rows and cols
        return combined
    else:
        return combined ,cols_drop
```

```python
def drop_test_cols(df_test,cols_drop):
    """ drop a list of cols that get from train set in test set
    """
    df_test=df_test.drop(columns=cols_drop)
    return df_test  
```

```python
def data_wrangling(combined,numerics, categoricals,trainset=True,clustering=True,cate_encoder=None):
    """ one hot encode categorical variables and fill null values for binary and numeric variables. If trainset is True,
    trainset encoder is returned for encoding testset and encoder will only be passed to data_wrangling function
    when processing testset.
    INPUT:
        combined: a dataframe with NaNs representing all unknowns
        numerics: a list of numeric variables
        categoricals: a list of categorical variables
        trainset: boolean, if dataset is trainset
        clustering: boolean, if function is used for clustering problem
        cate_encoder: pass a one hot encoder
    OUTPUT:
        df_final: a cleaned dataframe
    """
    if clustering:  # to check if the drop_rows_cols() function is used by clustering or classification
        # save these three columns for later clustering distribution visualization
        dataset_labels=combined[['dataset','ONLINE_PURCHASE', 'PRODUCT_GROUP']]
        # drop 'dataset','ONLINE_PURCHASE', 'PRODUCT_GROUP' columns
        combined=combined.drop(columns=['dataset','ONLINE_PURCHASE', 'PRODUCT_GROUP'])
    elif trainset:  # verify if it's test set or train set, if trainset, save response
        response=combined.RESPONSE
        combined=combined.drop(columns=['RESPONSE'])
        print(combined.shape)
    else:
        response=0

    
    # if there is not encoder passed, encode training set, otherwise, encode test set
    if cate_encoder == None: 
         # create one hot encoder for categorical variables
        cate_encoder = OneHotEncoder(handle_unknown='ignore')
        cate_dummies= cate_encoder.fit_transform(combined[[i for i in categoricals if i in combined.columns]].astype(str)).toarray()
        col_names=cate_encoder.get_feature_names([i for i in categoricals if i in combined.columns])
        cate_dummies_df=pd.DataFrame(cate_dummies,columns=col_names)  
        print(cate_dummies.shape)
        #     cate_dummies=pd.get_dummies(combined[[i for i in categoricals if i in combined.columns]].astype(str),drop_first=True)
    else:
        cate_dummies= cate_encoder.transform(combined[[i for i in categoricals if i in combined.columns]].astype(str)).toarray()
        col_names=cate_encoder.get_feature_names([i for i in categoricals if i in combined.columns])
        cate_dummies_df=pd.DataFrame(cate_dummies,columns=col_names)  
        
    
    # fill nan in numeric and ordinal variables with mode
    combined[numerics]=combined[[i for i in numerics if i in combined.columns]].apply(lambda x: x.fillna(x.mode()[0]),axis=0)
    print(combined.shape)
    # concatenate numeric variables and categorical variables
    df_final=pd.concat([combined[numerics].reset_index(drop=True),cate_dummies_df],axis=1)
    print(df_final.shape)

    if clustering:
        return df_final, dataset_labels    
    else:
        return df_final,response,cate_encoder
```

```python
def drop_correlated_cols(df_final,clustering=True,trainset=True,var_to_drop=None):
    """
    calculate pearson correlation for each variable pair and remove variables that are highly correlated
    with a correlation of 0.85. 

    INPUT:
        clustering: boolean, if it is function used for clustering
        trainset: boolean, if a trainset for classification is passed to this function
        df:a cleaned dataframe 
        var_to_drop: a list of variables that correlated to other variables that is gotten from trainset. Used for testset only
    OUTPUT:
        df_drop: a dataframe with no corrlated variables
    """
    
    # due to comupating time, pearson correlation for each pair of variables is calculated by using customers dataset, instead
    # of combination of azdias and customers
    if clustering:
        corr=df_final.iloc[azdias.shape[0]:,:].corr()
        corr_unstack=corr.abs().unstack()
        # get list of variable pairs with correlation higher than 0.85, and get the first variable of each pair to drop
        var_to_drop=[i[0] for i in corr_unstack[corr_unstack>0.85].index if i[0]!=i[1]]
        df_drop=df_final.drop(columns=var_to_drop)
        return df_drop
    elif trainset:
        corr=df_final.corr()
        corr_unstack=corr.abs().unstack()
        var_to_drop=[i[0] for i in corr_unstack[corr_unstack>0.85].index if i[0]!=i[1]]
        df_drop=df_final.drop(columns=var_to_drop)
        return df_drop,var_to_drop
    else:
        df_drop=df_final.drop(columns=var_to_drop)
        return df_drop
    
# df_drop=drop_correlated_cols(df_final,clustering=True,trainset=False,var_to_drop=None)

# print('dataset shape is {} after dropping correlated variables'.format(df_drop.shape))
```

### Dimensionality Reduction

```python
def scale_data(df_drop):
    """scale dataset with MinMax scaler"""
    # scale data
    scaler = MinMaxScaler()
    scaled_data=scaler.fit_transform(df_drop)
    
    return scaled_data

def plot_pca_component_variance(scaled_data):
    """ plot pca component variace figure to find the number of components that hits the point of diminishing return
    INPUT:
        scaled_data
    """
     # get accumulative variance and plot
    pca = PCA(n_components=scaled_data.shape[1],svd_solver='full')
    pca.fit(scaled_data)
    accum_variance=[np.sum(pca.explained_variance_ratio_[:idx+1]) for idx, val in enumerate(pca.explained_variance_ratio_) ]
    num_component=list(range(scaled_data.shape[1]))
    fig, ax = plt.subplots(figsize=[8,8])
    ax.bar(num_component, pca.explained_variance_ratio_,7,label='component variance ratio')
    ax.plot(num_component, accum_variance,'-m',label='accumulative variance ratio')
    ax.set_xlabel('# of component')
    ax.set_ylabel('Variance')
    ax.legend()
    ax.axvline(300,ls='--')
    ax.set_title('PAC component variance')

scaled_data=scale_data(df_drop)
plot_pca_component_variance(scaled_data)
```

To reduce reduce computational time of clustering, PCA is used to lower dimentionality. Cumulative sum of variance retained vs number of components is plotted along with variance for each individual component in a descending order. The plot shows that it hits diminishing returns when number of components reaches 300. Thus we select 300 as the number of components to keep for clustering

```python
def dimention_reduction(scaled_data,n_component):
    """perform dimentionality reduction with n_component determined from the pca component variance plot
    INPUT:
        scaled_data
        n_component: number of component to take
    OUTPUT:
        df_pca: a dataframe of n_components as columns, observations as rows
    """

    # PAC dimensionality reduction
    pca = PCA(n_components=n_component)
    pca_components=pca.fit_transform(scaled_data)
    df_pca=pd.DataFrame(pca_components)
    
    return df_pca

# reduce dimension of the dataset to 300
# df_pca=dimention_reduction(scaled_data,300)
print('dataset shape is {} after reducing dimensionality'.format(df_pca.shape))
```

```python
# start=time.time()
def determine_n_clusters(df_pca):
    """
    use kmeans elbow method to determine number of clusters.
    
    """
    clusters = []

    for i in range(1, 20):
        km = KMeans(n_clusters=i).fit(df_pca)
        clusters.append(km.inertia_)
        print('{} cluster done'.format(i))

    fig, ax = plt.subplots(figsize=[12, 8])
    ax.plot(list(range(1, 20)), clusters)
    ax.set_title('Searching for Elbow',size=13)
    ax.set_xlabel('Clusters',size=13)
    ax.set_ylabel('Inertia',size=13)

# end=time.time()
# print((end-start)/3600)

# determine_n_clusters(df_pca)
```

We use kmeans inertia to find out how well dataset is clustered under different number of clusters.It is calculated by measuring the distance between each data point and its centroid, squaring this distance, and summing these squares across one cluster. 

A good model is one with low inertia AND a low number of clusters (K). However, this is a tradeoff because as K increases, inertia decreases. To find the optimal K for a dataset, Elbow method is used to find the point where the decrease in inertia begins to slow. Unfortunately, in our case, there isn't a perfect elbow point. It appears that around cluster 6, decrease in inertia starts to slow. Just to be safe, we will use 10 clusters for analysis.




### KMeans Clustering

Now we are ready to cluster general population and the existing customers into clusters and visualize results

```python
def cluster_distribution_comparision(df_pca,n_clusters):
    """
    apply kmeans to cluster azdias and customers datasets into 10 clusters. Plot the percetage of observations distribution 
    in each cluster for each dataset and compare
    
    INPUT:
        df_pca: dataframe with reduced dimensions
    """
    # clustering with kmeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(df_pca)
    kmeans_labels=kmeans.labels_
    # create a dataframe with cluster labels and dataset labels
    clusters_df=pd.DataFrame({'kmeans_labels':kmeans_labels,'count':1})
    clusters_df=pd.concat([clusters_df,dataset_labels.reset_index(drop=True)],axis=1)
                          
    ##### online and offline distribution                     
    # get total number of observation in each dataset
    a=clusters_df[clusters_df.dataset=='azdias'].shape[0]
    b=clusters_df[clusters_df.dataset=='customers'].shape[0]
    # calculate percentage
    cluster_count=clusters_df.groupby(['kmeans_labels','dataset','ONLINE_PURCHASE']).count().unstack()
    cluster_count_percent=cluster_count/np.array([b,b,a,b,b,a])*100
    
    # get y values for bar chart
    azdias=[i for i in cluster_count_percent['count'][1000].values if ~np.isnan(i)]
    customers_online=[i for i in cluster_count_percent['count'][1].values if ~np.isnan(i)]
    customers_notonline=[i for i in cluster_count_percent['count'][0].values if ~np.isnan(i)]

    fig,ax=plt.subplots(figsize=[8,8])
    ax.bar(np.arange(1,11),azdias,0.3, label='azdias')
    ax.bar(np.arange(1,11)+0.3,customers_notonline,0.3, label='customers_offline',color='tab:orange')
    ax.bar(np.arange(1,11)+0.3,customers_online,0.3, label='customers_online',bottom=customers_notonline,color='gray')
    ax.set_xlabel('clusters')
    ax.set_ylabel('percent of observation in dataset')
    ax.legend()
    ax.set_title('cluster distribution comparision')
    
    
    ##### product group distribution
    cluster_count=clusters_df.groupby(['kmeans_labels','dataset','PRODUCT_GROUP']).count().unstack()
    cluster_count_percent=cluster_count/np.array([a,b,b,b,a,b,b,b])*100
    cluster_count_percent

    customers_cosmetic=[i for i in cluster_count_percent['count']['COSMETIC'].values if ~np.isnan(i)]
    customers_food=[i for i in cluster_count_percent['count']['FOOD'].values if ~np.isnan(i)]
    customers_food_cosmetic=[i for i in cluster_count_percent['count']['COSMETIC_AND_FOOD'].values if ~np.isnan(i)]

    fig,ax=plt.subplots(figsize=[8,8])
    ax.bar(np.arange(1,11),azdias,0.3, label='azdias')
    ax.bar(np.arange(1,11)+0.3,customers_cosmetic,0.3, label='customers_cosmetic',bottom=np.array(customers_food_cosmetic)+np.array(customers_food),color='m')
    ax.bar(np.arange(1,11)+0.3,customers_food,0.3, label='customers_food',bottom=customers_food_cosmetic,color='gray')
    ax.bar(np.arange(1,11)+0.3,customers_food_cosmetic,0.3, label='customers_food_cosmetic',color='tab:orange')
    ax.set_xlabel('clusters')
    ax.set_ylabel('percent of observation in dataset')
    ax.legend()
    ax.set_title('cluster distribution comparision')
    

# clustering dataset with 10 clusters and visualize the distribution of observations of each dataset in each cluster
# cluster_distribution_comparision(df_pca,n_clusters=10)
```

### Build ML pipeline

As shown in the above bar chart, there is a large output class imbalance, where most individuals did not respond to the mailout. Thus, predicting individual classes and using accuracy does not seem to be an appropriate performance evaluation method. Instead, **ROC_AUC is used as evaluation metric**. The exact values of the "RESPONSE" column do not matter as much: only that the higher values try to capture as many of the actual customers as possible, early in the ROC curve sweep.

```python
def build_model(pipeline,search_spaces):
    """ build machine learning pipeline and use BayeSearchCV for hyperparameter tunning
    
        INPUT:
            pipeline: ML pipeline model
            search_spaces: a dictionary of parameters for baye search to tune
        OUPUT:
            model: tuned model
    """
 
    model = BayesSearchCV(
         pipeline,
         search_spaces=search_spaces,
         n_iter=50,scoring='roc_auc',
         random_state=0,cv=2,return_train_score=True,verbose=3
         )
    
    return model
```

```python
def plot_train_test_score(model,model_name):
    """ plot the changes of model train and test roc_auc scores in BayeSearchCV iterations
    """
    print('best estimator for {}: '.format(model_name),model.best_params_)
    print('\nbest score: ',model.best_score_)
    
    fig,ax=plt.subplots(figsize=[7,7])
    ax.plot(model.cv_results_[ 'mean_test_score'],label='test roc_auc score')
    ax.plot(model.cv_results_[ 'mean_train_score'],label='train roc_auc score')
    plt.xlabel('num of iteration')
    plt.ylabel('roc_auc score')
    plt.title('{} train vs test roc_auc score'.format(model_name))
    plt.legend()
              
```
